```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个9x9的数组，初始值都为0
table = np.zeros((9, 9))

# 将以中心位置为中心的5x5范围内的值都设为1
table[2:7, 2:7] = 1

# 绘制表格
fig, ax = plt.subplots()
ax.axis('off')

# 根据table中的值设置单元格颜色
cell_colors = [['blue' if table[i, j] == 1 else 'white' for j in range(9)] for i in range(9)]
# 把位置（4，4）的颜色变为红色
cell_colors[4][4] = 'red'

ax.table(cellText=table, cellLoc='center', loc='center', bbox=[0, 0, 1, 1], cellColours=cell_colors)

# 显示图表
plt.show()
```
